<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
   <p>The sieve data is plotted in the figure below. Note that the most typical probability density function for grain diameters is log-normal. Therefore, we should create our plot should have a log-scale on the axis corresponding to grain diameter.</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
seddata = pd.DataFrame(
    [[0.074,  0.07],
     [0.125,  0.33],
     [0.246,  1.70],
     [0.351,  6.20],
     [0.495, 19.00],
     [0.701, 37.30],
     [0.991, 60.50],
     [1.400, 79.40],
     [1.980, 90.40],
     [3.960, 99.30]], 
    columns =['D','Pf'])
flowdata = pd.DataFrame(
    [[0.16, 0.21, np.nan],
     [0.18, 0.23, np.nan],
     [0.22, 0.32, 0.004],
     [0.25, 0.40, 0.006],
     [0.27, 0.48, 0.007],
     [0.29, 0.52, 0.009],
     [0.32, 0.60, 0.012],
     [0.40, 0.87, 0.030],
     [0.43, 1.10, 0.039],
     [0.60, 1.50, np.nan]], 
    columns = ['H','q','qbw'])
seddata
```

   <p>Now we can plot the data using matplotlib (see code below).</p>

```python
# set some figure formats
%config InlineBackend.figure_format = 'svg' 
plt.rcParams.update({'font.size': 12})
plt.rcParams['font.family']='STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'stix'
# make the plot
fig,ax = plt.subplots()
ax.semilogx(seddata['D'],seddata['Pf'],'.-')
ax.set_xticks(2.**np.arange(-4,3,1))
ax.xaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.minorticks_off()
ax.tick_params(axis='both', direction='in',
               top=True, right=True, length=4)
ax.set_ylabel('Percent finer, $P_f$ (%)')
ax.set_xlabel('Grain diameter, $D$ (mm)')
plt.show()
```

<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
    <p>Characteristic grain diameters, $D_{50}$ and $D_{90}$, can be determined by interpolating the sieve data (in $\phi$-space) for the 50<sup>th</sup> and 90<sup>th</sup> percentiles, respectively.</p>
    <p>Because of the log-normal nature of grain size distributions, the interpolation should be on log-transformed diameters, which is the same as the $\phi$-scale.</p>
    <p>It is possible to do this visually by zooming in on the plot shown above, but it is better to numerically interpolate using the data in the sieve analysis table.</p>
    <p>Recall that in order to linearly interpolate for a value ($x$,$y$) between two other data points ($x_a$,$y_a$) and ($x_b$,$y_b$) the appropriate relation is </p>
    <p>$$y = y_a+(y_b-y_a)\dfrac{x-x_a}{x_b-x_a}$$</p>
    <p>In this case, $x_a$ and $x_b$ correspond to percent finer values falling directly below and above the percentile being solved for, and $y_a$ and $y_b$ correspond to the $\phi$ values for those percentiles. Or,</p>
    <p>$$\phi_X = \phi_a+(\phi_b-\phi_a)\dfrac{X-P_{f,a}}{P_{f,b}-P_{f,a}}$$</p>
    <p>To determine the median grain diameter, or $D_{50}$ we can use </p>
    <p>$$\phi_{50}=-\log_2(0.701\mathrm{\,mm})+\left[-\log_2(0.991\mathrm{\,mm})+\log_2(0.701\mathrm{\,mm})\right]\dfrac{50\%-37.3\%}{60.5\%-37.3\%}$$</p>
    <p>$$\phi_{50}=0.239$$</p>
    <p>Recall that $2^{-\phi}$. Therefore,</p>
    <p>$$D_{50}=2^{-\phi_{50}}=2^{-0.239}$$</p>
    <p>$$\boxed{D_{50}=0.847\mathrm{\,mm}}$$</p>
    <p>We can do the same thing for the $D_{90}$.</p>
    <p>$$\phi_{90}=-\log_2(1.4\mathrm{\,mm})+\left[-\log_2(1.98\mathrm{\,mm})+\log_2(1.4\mathrm{\,mm})\right]\dfrac{90\%-79.4\%}{90.4\%-79.4\%}$$</p>
    <p>$$\phi_{90}=-0.967$$</p>
    <p>$$D_{90}=2^{-\phi_{90}}=2^{0.967}$$</p>
    <p>$$\boxed{D_{90}=1.955\mathrm{\,mm}}$$</p>
</div>

<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
        <p>In the code, we will add a column of $\phi$ values to the table corresponding to each $D$ value. Then, we will interpolate using that column and the percent finer, $P_f$, column. Interpolation is done through the numpy <span style="font-family:courier;">interp</span> command (see code below).</p>
</div>

```python
seddata['phi'] = -np.log2(seddata['D'])
Pfx = [50,90]
Dx = 2.**-np.interp(Pfx,seddata['Pf'],seddata['phi'])
print("D50 = %5.3f mm" % Dx[0])
print("D90 = %5.3f mm" % Dx[1])
```

</div>

```python
ax.plot(Dx,Pfx, '*w', ms=20, mec='k', mew=1.5)
fig
```

<li style="font-family: helvetica; margin-bottom: 10px;"><span style="font-size: 12pt; font-family: helvetica;"><strong>Fall Velocity</strong>&nbsp; &nbsp; <span style="font-family: 'times new roman', times;">Determine the fall velocity for <em>D</em><sub>50</sub>, assuming that the flow around the particle is:&nbsp;</span> </span>
<ol style="list-style-type: lower-alpha; font-family: helvetica;">
<li><span style="font-size: 12pt; font-family: times new roman, times;">laminar, and</span></li>
<li style="font-family: helvetica;"><span style="font-size: 12pt; font-family: 'times new roman', times;">turbulent</span> <br style="font-family: 'times new roman', times;"><span style="font-size: 12pt; font-family: 'times new roman', times;">Also compute the relevant Reynolds number and state which of the two answers (a or b) is appropriate.</span></li>
</ol>
</li>
</ol>

<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
    <p>Laminar fall velocity, $w_{0,\mathrm{lam}}$, for the $D_{50}$ grain size can be determined using</p>
    <p>$$\begin{align}
        w_{0,\mathrm{lam}}&=\dfrac{1}{18}\dfrac{(s-1)gD_{50}^2}{\nu}\\
    \end{align}$$</p>
    <p>where $s$ is the specific gravity of sediment ($s=2.65$), $g$ is gravitational acceleration ($g=32.2$ ft/s<sup>2</sup>), and $\nu$ is the kinematic viscosity ($\nu=1.08\times10^{-5}$ ft<sup>2</sup>/s). Therefore,</p>
    <p>$$
        w_{0,\mathrm{lam}}=\dfrac{1}{18}\dfrac{\left(2.65-1\right)\times32.2\mathrm{\,ft}/\mathrm{s}^2\times\left(0.847\mathrm{\,mm}\times\dfrac{1\mathrm{\,ft}}{304.8\mathrm{\,mm}}\right)^2}{1.08\times10^{-5}\mathrm{\,ft}^2/\mathrm{s}}\\
    $$</p>
    <p>$$\boxed{w_{0,\mathrm{lam}}=2.112\mathrm{\,ft}/\mathrm{s}}\\
        \mathrm{or}\\
        \boxed{w_{0,\mathrm{lam}}=0.645\mathrm{\,m}/\mathrm{s}}$$</p>
</div>

```python
s = 2.65 # specific gravity of sediment
g = 32.2 # ft/s2, gravitational acceleration
nu = 1.08e-5 # ft2/s, kinematic viscosity of water
w0lam = 1/18*(s-1)*g*(Dx[0]/304.8)**2/nu
print("w0,lam = %5.3f ft/s" % w0lam)
```

<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
    <p>For turbulent fall velocity, we first need to calculate the dimensionless grain size</p>
    <p>$$D_* = D\left[\dfrac{\left(s-1\right)g}{\nu^2}\right]^{1/3}$$</p>
    <p>$$D_* = 0.847\mathrm{\,mm}\times\dfrac{1\mathrm{\,ft}}{304.8\mathrm{\,mm}}\times\left[\dfrac{\left(2.65-1\right)\times32.2\mathrm{\,ft}/\mathrm{s}^2}{\left(1.08\times10^{-5}\mathrm{\,ft}^2/\mathrm{s}\right)^2}\right]^{1/3}$$</p>
    <p>$$D_* = 21.39$$</p>
    <p>From this, we can calculate the particle Reynolds number using our relation for natural grains</p>
    <p>$$\mathrm{Re}_p=8\left[\sqrt{1+0.0139D_*^3}-1\right]$$</p>
    <p>$$\mathrm{Re}_p=8\left[\sqrt{1+0.0139\times\left(21.39\right)^3}-1\right]$$</p>
    <p>$$\mathrm{Re}_p=85.64$$</p>
    <p>Finally, we can use the definition of the particle Reynolds to be back-calculate turbulent fall velocity.</p>
    <p>$$\mathrm{Re}_p=\dfrac{w_0D}{\nu}\Rightarrow w_{0,\mathrm{turb}}=\mathrm{Re}_p\dfrac{\nu}{D_{50}}$$</p>
    <p>$$w_{0,\mathrm{turb}}=85.64\times\dfrac{1.08\times10^{-5}\mathrm{\,ft}^2/\mathrm{s}}{0.847\mathrm{\,mm}\times\dfrac{1\mathrm{\,ft}}{304.8\mathrm{\,mm}}}$$</p>
    <p>$$\boxed{w_{0,\mathrm{turb}}=0.333\mathrm{\,ft}/\mathrm{s}}\\
        \mathrm{or}\\
        \boxed{w_{0,\mathrm{turb}}=0.101\mathrm{\,m}/\mathrm{s}}$$</p>
</div>

```python
Dstar = Dx[0]/304.8*(g*(s-1)/nu**2)**(1/3)
print("D* = %5.2f" % Dstar)
Rep = 8*(np.sqrt(1+0.0139*Dstar**3)-1)
print("Rep = %5.2f" % Rep)
w0turb = Rep*nu/Dx[0]*304.8
print("w0,turb = %5.3f ft/s" % w0turb)
```

<li style="font-size: 12pt; font-family: helvetica; margin-bottom: 10px;"><span style="font-size: 12pt; font-family: helvetica;"><strong>Laminar Fall Velocity</strong>&nbsp; &nbsp; </span> <span style="font-family: times new roman, times;">Determine the diameter of the largest particle for which settling can still be considered laminar. Note that this question has nothing to do with the data in the table above.</span></li>
</ol>

<div style="margin-left: 50px; color: #0000FF; font-family:'times new roman', times; font-size:12pt;">
    <p>The largest particle, $D_\mathrm{max}$, for which settling can still be considered laminar will correspond to a particle Reynolds number of $\mathrm{Re}_p=1$.</p>
    <p>$$\mathrm{Re}_p=1=\dfrac{w_0D_\mathrm{max}}{\nu}\quad\Rightarrow\quad w_0=\dfrac{\nu}{D_\mathrm{max}}$$</p>
    <p>Combining this with the equation for laminar settling velocity, we can rearrange to solve for $D_\mathrm{max}$.</p>
    <p>$$w_0=\dfrac{\nu}{D_\mathrm{max}}=\dfrac{1}{18}\dfrac{(s-1)gD_\mathrm{max}^2}{\nu}$$</p>
    <p>$$\therefore$$</p>
    <p>$$D_\mathrm{max}=\left[\dfrac{18\nu^2}{\left(s-1\right)g}\right]^{1/3}$$</p>
    <p>$$D_\mathrm{max}=\left[\dfrac{18\times\left(1.08\times10^{-5}\mathrm{\,ft}^2/\mathrm{s}\right)^2}{\left(2.65-1\right)\times32.2\mathrm{\,ft}/\mathrm{s}^2}\right]^{1/3}$$</p>
    <p>$$\boxed{D_\mathrm{max}=0.00034\mathrm{\,ft}}\\
        \mathrm{or}\\
        \boxed{D_\mathrm{max}=0.104\mathrm{\,mm}}$$</p>
</div>

```python
Dmax = (18*nu**2/(s-1)/g)**(1/3)
print("Dmax = %7.5f ft" % Dmax)
print("Dmax = %5.3f mm" % (Dmax*304.8))
```
