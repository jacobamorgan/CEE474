```python
#########################################################
#########################################################
# Lecture 03 - Specific Energy
# Creating for CEE 474 / CEWA 574, Spring 2021
# By Jacob A. Morgan, Ph.D.
#    University of Washington
#########################################################
#########################################################
import numpy as np
from matplotlib import pyplot as plt
#########################################################
# INPUT/GIVENS
g = 981 # gravity (cm/s^2)
Q = 1000 # volumetric flow rate (cm^3/s)
b = 5 # channel width (cm)
deltaz = 1 # "bump" height (cm)
q = Q/b # unit discharge (cm^2/s)
#########################################################
# RANGE FOR PLOT
h = np.linspace(1,15.0,101) # flow depths (cm)
E = h+q**2/2/g/h**2 # specific energies (cm)
#########################################################
# UPSTREAM SECTION
h1 = np.array([8., 1.76357949]) # upstream depth (cm)
E1 = h1+q**2/2/g/h1**2 # upstream specific energy (cm)
#########################################################
# DOWNSTREAM SECTION
E2 = np.mean(E1)-deltaz # downstream specific energy (cm)
h2 = np.roots([1,-E2,0,q**2/2/g]) # downstream depth (cm)
h2 = h2[h2>0] # remove non-physical downstream depth
#########################################################
# CRITICAL FLOW
hc = (q**2/g)**(1/3) # critical depth (cm)
Ec = 3/2*hc # minimum specific energy (cm)
#########################################################
# PLOT
fig, ax = plt.subplots(figsize=(4,4), dpi= 100)
plt.plot([0,15],[0,15],'--k')
plt.plot(E,h,lw=4)
plt.plot(E1,[0,15],':r')
plt.plot(E1,h1,'or',ms=10)
plt.plot([E2,E2],[0,15],':r')
plt.plot([E2,E2],h2,'or',ms=10,mfc='w')
plt.plot(Ec,hc,'ok',ms=10)
plt.ylabel(r'Flow depth, $h$ (cm)')
plt.xlabel(r'Specific energy, $E$ (cm)')
plt.ylim([0,15]), plt.xlim([0,15])
plt.yticks(range(0,15,2)), plt.xticks(range(0,15,2))
ax.set_aspect('equal', adjustable='box')
plt.show()
```
