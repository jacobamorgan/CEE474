# The Shields Diagram

The Shields diagram shows the relationship between the non-dimensional shear stress and the boundary Reynolds number at the threshold of particle motion.

Shear stress is nondimensionalized by product of the relative submerged specfic weight of the particles and the particle diameter. This parameter is often referred to as the Shields parameter, Shields stress, Shields number, mobility parameter, or simply the nondimensional shear stress and is commonly denoted by $\theta$ or $\tau_*$.

$$\tau_* = \dfrac{\tau_0}{\left(\gamma_s-\gamma\right)D} = \dfrac{\tau_0}{\left(\rho_s-\rho\right)gD} = \dfrac{u_*^2}{\left(s-1\right)gD}$$

where $\tau_0$ is the boundary shear stress, $\gamma_s$ is the specific weight of the particles, $\gamma$ is the specific weight of the fluid, $D$ is the particle diameter, $\rho_s$ is the density of the particles, $\rho$ is the density of the fluid, $g$ is gravitational acceleration, $u_*$ is the shear velocity (defined as $u_*=\sqrt{\tau_0/\rho}$), and $s$ is the specific gravity of the particles (defined as $\rho_s/\rho$).

By multiplying both the numerator and the denominator by $D^2$, we can easily see that the Shields number represents the ratio of the fluid force due to shear stress on a particle to the submerged weight of the particle.

The boundary Reynolds number, denoted $\mathrm{Re}_*$, is related to the shear velocity and particle diameter.

$$\mathrm{Re}_* = \dfrac{\rho u_*D}{\mu}=\dfrac{u_*D}{\nu}=\dfrac{D}{\delta}$$

where $\mu$ is the dynamic viscosity of the fluid, $\nu$ is the kinematic viscosity of the fluid, and $\delta$ is a boundary layer length scale.

In Albert Shield's dissertation (Shields, 1936a,b), he presented data from his own experiments as well as data compiled from other sources, to show the relationship between the critical tractive force, or critical shear stress, and the thickness of the viscous sublayer.
```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams.update({'font.size': 18})
plt.rcParams['font.family']=['STIXGeneral','Times New Roman','DejaVu Serif']
mpl.rcParams['mathtext.fontset'] = 'stix'

data = np.genfromtxt('shields_original_data.csv', delimiter=',', skip_header=1,
                     dtype=[('tau_star_c','f8'),
                            ('Re_star_c','f8'),
                            ('D50_mm','f8'),
                            ('rho_s','f8'),
                            ('reference','S20')])
refs = np.unique(data['reference'])

dstar = np.power(10,np.linspace(0.5,2.333,100))
# Original Shields Curve
theta_shields = 0.24/dstar+0.055*(1-np.exp(-0.02*dstar))

fig, ax = plt.subplots(figsize=(10,5), dpi= 100, facecolor='w', edgecolor='k')
ax.set_prop_cycle(color=['blue', 'purple', 'green', 'red', 'orange'],
                  marker=['v', '^', 'X', 'o', 'P'])
colors = ('blue', 'purple', 'green', 'red', 'orange')
shapes = ('v', '^', 'X', 'o','P')
plt.plot((theta_shields*dstar**3)**(1/2),theta_shields, linewidth=3, marker='', color='black', linestyle='--',
         label=r"$\tau_{*c}$"+", Soulsby & Whitehouse (1997)")#"\n"+r"$\theta_c=\frac{0.24}{d_*}+0.055\left(1-e^{-0.02d_*}\right)$")
for ind, i in enumerate(refs):
    plt.plot(data['Re_star_c'][data['reference']==i],data['tau_star_c'][data['reference']==i],linestyle='',
             markersize=17,color=colors[ind],marker=shapes[ind],markeredgecolor='k',markeredgewidth=2,label=repr(i)[3:-1])
    plt.plot(data['Re_star_c'][data['reference']==i],data['tau_star_c'][data['reference']==i],linestyle='',
             markersize=15,color=colors[ind],marker=shapes[ind])
ax.loglog()
plt.ylim((0.01,0.25))
plt.ylabel("Shields Number\n"+r"$\tau_*=\dfrac{\tau_0}{\left(\rho_s-\rho\right)gD} = \dfrac{u_*^2}{\left(s-1\right)gD}$",
          fontsize=20)
plt.xlabel("Boundary Reynolds Number\n"+r"$\mathrm{Re}_*=\dfrac{\rho u_*D}{\mu}=\dfrac{u_*D}{\nu}=\dfrac{D}{\delta}$",
          fontsize=20)
plt.legend(fontsize=17, ncol=2, edgecolor='black', fancybox=False)
ax.tick_params(which="both", right=True, top=True, direction="in", length=5)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))
```

You will notice that Shields actually did not delineate a single curve, but rather showed a shaded region. Following the first appearance of Shields' diagram in the English-language literature (Rouse, 1939), it became customary to show crticial shear stress as a single line rather than a shaded region.
